```python
import pandas as pd
import numpy  as np 
import os
for dirname, _, filenames in os.walk('C:/Users/DELL/Desktop/titanic data'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
tf = pd.read_csv("C:/Users/DELL/Desktop/titanic data/train.csv")
tf.drop(['Ticket','Cabin'],axis=1,inplace=True)
tf.head()
```

```python
gender_mapping={'male':0,'female':1}
class_map = {'C':0,'S':1,'Q':2}
tf['Embarked'] = tf['Embarked'].map(class_map)
tf['Sex']  = tf['Sex'].map(gender_mapping)
tf['Age'] = tf['Age'].fillna(tf['Age'].median())
tf['Embarked'].fillna(0,inplace=True)
tf['Fare']=tf['Fare']/20
tf['Age'] = tf['Age']/10
tf.head()
```

```python
for i in range(tf.shape[0]):
    if("Master" in str(tf.loc[i,"Name"])):
        tf.loc[i,"Name"] = 0
    elif("Mr." in str(tf.loc[i,"Name"])):
        tf.loc[i,"Name"] = 0.5
    elif("Mrs." in str(tf.loc[i,"Name"])):
        tf.loc[i,"Name"] = 1
    elif("Miss." in str(tf.loc[i,"Name"])):
        tf.loc[i,"Name"] = 1.5
    elif("Rev." in str(tf.loc[i,"Name"])):
        tf.loc[i,"Name"] = 2
    elif("Dr." in str(tf.loc[i,"Name"])):
        tf.loc[i,"Name"] = 3.5
    elif("Sir" in str(tf.loc[i,"Name"])):
        tf.loc[i,"Name"] = 4
    else:
        tf.loc[i,"Name"] = 4.5
tf["Name"].value_counts()
```

```python
m= [[],[]]
l=[[],[],[]]
for i in range(tf.shape[0]):
    if(tf.loc[i,"Pclass"]==1):
        l[0].append(1)
        l[1].append(0)
        l[2].append(0)
    elif(tf.loc[i,"Pclass"]==2):
        l[0].append(0)
        l[1].append(1)
        l[2].append(0)
    else:
        l[0].append(0)
        l[1].append(0)
        l[2].append(1)    
    if(tf.loc[i,"Sex"]==1):
        m[0].append(1)
        m[1].append(0)
    else:
        m[0].append(0)
        m[1].append(1)
tf["p1"]=l[0]
tf["p2"]=l[1]
tf["p3"]=l[2]
tf["male"]=m[0]
tf["female"]=m[1]
tf.head()
```

```python
from sklearn import preprocessing
from sklearn import utils
from sklearn.linear_model import LogisticRegression as dre
#from sklearn.ensemble import RandomForestClassifier as dre
y = tf["Survived"]
features = ["male","female","Parch","Age","p1","p2","p3","SibSp","Embarked","Fare","Name"]
X = pd.get_dummies(tf[features])
#model = dre(n_estimators=200,max_depth=40,n_jobs=-1)
model = dre(n_jobs=-1,max_iter=10000)
model.fit(X, y)
predictions = model.predict(X)
#output = pd.DataFrame({'PassengerId': train_data.PassengerId, 'Survived': predictions})
#output.to_csv('my_submission.csv', index=False)
#print("Your submission was successfully saved!")
from sklearn.metrics import f1_score
print((f1_score(tf['Survived'],predictions))*100)
```
